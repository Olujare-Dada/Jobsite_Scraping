```python
def remove_raw_data_from_path(file_path):
    
    if os.path.exists(file_path):
        os.remove(file_path)
        print(f"File '{file_path}' has been deleted.")
    
    else:
        print(f"File '{file_path}' does not exist.")
```

```python
from dotenv import load_dotenv
from azure.identity import DefaultAzureCredential
from azure.storage.blob import BlobServiceClient, BlobClient, ContainerClient
from azure.core.exceptions import ServiceRequestError, AzureError, ClientAuthenticationError, DecodeError, DeserializationError, HttpResponseError, IncompleteReadError, ResourceNotFoundError, ResourceExistsError, SerializationError, ServiceRequestError, ServiceRequestTimeoutError, ServiceResponseError, ServiceResponseTimeoutError, StreamClosedError, StreamConsumedError, TooManyRedirectsError


list_of_azure_errors = [ServiceRequestError, AzureError, ClientAuthenticationError, DecodeError, DeserializationError, HttpResponseError, IncompleteReadError, ResourceNotFoundError, ResourceExistsError, SerializationError, ServiceRequestError, ServiceRequestTimeoutError, ServiceResponseError, ServiceResponseTimeoutError, StreamClosedError, StreamConsumedError, TooManyRedirectsError]


#load_dotenv("azure_client_credentials.env")
# Acquire a credential object
token_credential = DefaultAzureCredential()

blob_service_client = BlobServiceClient(
        account_url=f"https://{storage_account_name}.blob.core.windows.net",
        credential=token_credential)




def upload_blob_file(blob_service_client: BlobServiceClient, container_name: str, local_file_path: str, blob_name: str):
    # Get the container client
    container_client = blob_service_client.get_container_client(container_name)

    # Specify the path to the local file you want to upload
    with open(local_file_path, "rb") as data:
        # Upload the file to the container
        blob_client = container_client.upload_blob(name=blob_name, data=data, overwrite=True)

    # Return the URL of the uploaded blob
    return blob_client.url


list_of_files = os.listdir(os.getcwd())

if "Jobberman_data_raw.parquet" in list_of_files:
    
    curr_time = datetime.now()
    datetime_val = curr_time.strftime("%Y-%m-%d %H:%M:%S")
    
    current_try = 1
    max_retries = 3
    
    while max_retries > 0:
        print(f"Number of tries: {current_try}")
        try:
            upload_blob_file(blob_service_client, azure_container_name[0], f"./Jobberman_data_raw.parquet", 
                             f"Jobberman_data_raw_{datetime_val}.parquet")
            print("Jobberman_data_raw.parquet file has been uploaded to Azure container. Deleting from local system...")
            break
        
        except list_of_azure_errors as e:
            print(f"An error occurred!\nPython error: {e}")
            
        
        max_retries -= 1
        current_try += 1

        

    remove_raw_data_from_path("./Jobberman_data_raw.parquet")
    
    
if "Jobberman_data_raw.csv" in list_of_files:
    remove_raw_data_from_path("Jobberman_data_raw.csv")
```
